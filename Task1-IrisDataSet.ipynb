```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='petal_length',y='petal_width',hue='species',s=70)
plt.title("Scatter Plot of Petal Length vs. Petal Width")
plt.xlabel("Petal Length(cm)")
plt.ylabel("Petal width(cm)")
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.legend(title='species')
plt.show()
```

```python
plt.figure(figsize=(8,6)) #size of the output
#Making histogram
sns.histplot(data=df,x='sepal_length', kde = True, bins=15)
plt.title('Distribution of sepal length')
plt.xlabel('Sepal length(cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
#Making Boxplot
sns.boxplot(data=df,x='species', y="sepal_width", hue='species', palette='viridis')
plt.title('Box plot of sepal width by species')
plt.xlabel('Species')
plt.ylabel('Species width (cm)')
plt.show()
```
